```python
# import usual things
import pandas as pd
import bqplot
import numpy as np
import traitlets
import ipywidgets
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import the data file
building = pd.read_csv('building_inventory.csv')
```

```python
building
```

```python
#data transofrmation
list_ = []

for i in range(len(building['Congress Dist'].unique())):
    df = building.loc[building['Congress Dist'] == building['Congress Dist'].unique()[i]]
    for j in range(len(building['Agency Name'].unique())):
        sum_ = df.loc[df['Agency Name'] == building['Agency Name'].unique()[j], 'Square Footage'].sum()
        list_.append([building['Congress Dist'].unique()[i],building['Agency Name'].unique()[j],sum_])
list_
```

```python
#conversion of the transformed data back to a dataframe
building_new = pd.DataFrame(list_,columns = ['Congress Dist','Agency Name','Square Footage'])
building_new
```

```python
#transforming the square footage column
square_footage = building_new['Square Footage'].to_numpy()
square_footage = square_footage.reshape(19,35)
square_footage
```

```python
#grid heat map
from bqplot import pyplot
# 2. scales - linear scales
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale()
c_sc = bqplot.ColorScale()

# 3. axis
ax_x = bqplot.Axis(scale = x_sc, label='Agency Name')
ax_y = bqplot.Axis(scale = y_sc, orientation='vertical', label='Congress Dist')
ax_col = bqplot.ColorAxis(scale = c_sc, orientation='vertical', side='right') 
# 4. Marks
heat_map = bqplot.GridHeatMap(color = square_footage, #change
                              row = building_new['Congress Dist'].unique(), #change
                              column = building_new['Agency Name'].unique(), #change
                              scales = {'color':c_sc, 'row':y_sc, 'column':x_sc},
                             interactions={'click':'select'},
                             anchor_style={'fill':'blue'})
                             #selected_style={'opacity':1.0},
                             #unselected_style={'opacity':0.8})

mySelectedLabel = ipywidgets.Label()
def on_selected(change): # already modified for our 3d data
    if len(change['owner'].selected) == 1:
        i, j = change['owner'].selected[0]
        v = square_footage[i,j] # this is a 2d array!!
        mySelectedLabel.value = 'Weighted binning (log(duration in sec)) = ' + str(v)
        ##### Add in what changes for our histogram
        ####hist.sample = data3d[i,j]
# make sure we are observing for changes in our heatmap
heat_map.observe(on_selected, 'selected')
        
# figure!
fig = bqplot.Figure(marks=[heat_map], axes=[ax_col, ax_y, ax_x])
#fig

# dashboard with my label
myDashboard = ipywidgets.VBox([mySelectedLabel, fig])
myDashboard
```

```python
#preprocessing for creation of line plot
building_line = building.groupby('Year Acquired').sum().reset_index()
building_l = building_line[building_line['Year Acquired'] != 0]
building_l
```

```python
#further preprocessing
square_foot = building_l['Square Footage'].to_numpy()
square_foot
```

```python
# line plot creation
fig_l = bqplot.pyplot.figure(title='Square Footage Acquired Per Year')
axes_options = {'x': {'label': 'Year Acquired', 'tick_format': '0.0f'},
                'y': {'label': 'Square Footage', 'tick_format': '0.0f'}}
```

```python
line = bqplot.pyplot.plot(x =building_l['Year Acquired'], y = square_foot, axes_options= axes_options)
```

# Things to think about

Keeping the x and y axes as static on the line plot does not seem like an ideal thing to do because their ranges are dependent on the cell selected in the Grid Heat Map. Static ranges might either lead to missing values if the ranges don't consider the absolute minimum and the absolute maximum values or lead to very small and concentrated visualizations when the ranges include all the values that can be wtinessed.

The style can definitely be changed based on how the dashboard creator wants to view the various components of the dashboard and the interactivity between these components. 

```python
#5. interactivity
mySelectedLabel = ipywidgets.Label()
def on_selected(change): # already modified for our 3d data
    if len(change['owner'].selected) == 1:
        i, j = change['owner'].selected[0]
        v = square_footage[i,j] # this is a 2d array!!
        mySelectedLabel.value = 'Total Square Footage: ' + str(v)
        line.y = square_foot[i:j]
```

```python
#line plot
fig_l = bqplot.pyplot.figure(title='Square Footage Acquired Per Year')
axes_options = {'x': {'label': 'Year Acquired', 'tick_format': '0.0f'},
                'y': {'label': 'Square Footage', 'tick_format': '0.0f'}}
line = bqplot.pyplot.plot(x =building_l['Year Acquired'], y = square_foot, axes_options= axes_options)
```

```python
# mess with some layout parameters (look a little like CSS stylings)
fig.layout.min_width='500px'
fig_l.layout.min_width='500px'

figures = ipywidgets.HBox([fig, fig_l]) # figures side-by-side

# label on top
myDashboard = ipywidgets.VBox([mySelectedLabel, figures])
myDashboard
```

# Data Transformation and Rescaling in the plots

**1. For the Grid Heat Map, a lot of data transformations were done which are mentioned below.**

First, unique values for Congressional District and Agency Names were compared to each other to see which Agency Names were a part of a specific Congressional District and which weren't. This helped in determining a particular value for the Square Footage which was greater than zero for Congressional Districts constituting specific Agencies. In case of agencies that were not a part of a specific Congressional District, the Square Footage was totalled to zero. 

All these changes were stored in the form of a list of list with the Congressional District, Agency Names and Sum of Square Footage values being stored in the same order. This list of list was then converted to a pandas dataframe with their original headers. 

Now, the Sum of Square Footage values were converted to a 2D Numpy array and this array was reshaped to correctly fit the Grid Heat Map with the 'Agency Names' as columns and 'Congressional District' as rows.



**2. For the Line plot, following data tansformations were done.**

First, groupby operation was performed on the 'Year Acquired' column of the original dataframe to obtain the Total Sum of Square Footage values for each 'Year Acquired' by using the SUM aggregate function.

Further, the row with 'Year Acquired' equivalent to zero was removed from the dataframe to eliminate outliers.

Then, the dataframe column represting the Total Sum of Square Footage values was converted to a numpy array to enable linking of the heat map with the line plot.
